**4 - Cerca llibreries de logs en altres llenguatjes (al menys 2, i identifica cóm resolen les següents característiques típiques d’un sistema de logging.**

| **Característica** | **Python (Logging Module)** | **Java (Logback)** |
|--------------------|-----------------------------|--------------------|
| **Nom de la llibreria** | `logging` | `Logback` |
| **És nativa del llenguatge?** | Sí | No |
| **URL per descargar-se la llibreria** | [Documentació oficial](https://docs.python.org/3/library/logging.html) | [Logback oficial](https://logback.qos.ch/) |
| **Inicialització de l’objecte de logger** | `import logging`<br>`logger = logging.getLogger('my_logger')` | `Logger logger = LoggerFactory.getLogger(MyClass.class)` |
| **Nivells de log disponibles** | `DEBUG`, `INFO`, `WARNING`, `ERROR`, `CRITICAL` | `TRACE`, `DEBUG`, `INFO`, `WARN`, `ERROR`, `FATAL` |
| **Mètode per fer log** | `logger.debug()`, `logger.info()`, `logger.warning()`, etc. | `logger.debug()`, `logger.info()`, `logger.error()`, etc. |
| **Tipus de manegadors (pantalla, fitxer…) Identificar els seus noms a la API** | `StreamHandler` (pantalla), `FileHandler` (fitxer) | `ConsoleAppender`, `FileAppender` |
| **Opcions de format** | Configurable via `basicConfig()`<br>(ex. `format='%(levelname)s:%(message)s'`) | Es defineix en un fitxer XML<br>(ex. `<PatternLayout pattern='%d{HH:mm:ss} %-5level [%logger] - %msg%n' />`) |

### Descripció:
1. **Python (Logging Module)**:
   - **Nativa**: El mòdul `logging` és part de la biblioteca estàndard de Python, la qual cosa facilita la seva utilització sense necessitat d'instal·lar llibreries addicionals.
   - **Flexibilitat**: Permet definir formats personalitzats, nivells de log, i gestionar múltiples manegadors de logs com a fitxers o pantalles a través d'un codi relativament simple&#8203;:contentReference[oaicite:0]{index=0}.

2. **Java (Logback)**:
   - **No nativa**: Logback és una llibreria externa àmpliament utilitzada en Java per gestionar els logs. No ve instal·lada per defecte, però és molt popular.
   - **Configuració XML**: A través d'un fitxer de configuració XML, els usuaris poden personalitzar fàcilment el format i els manegadors (com `ConsoleAppender` i `FileAppender`) per gestionar on i com es registren els logs&#8203;:contentReference[oaicite:1]{index=1}&#8203;:contentReference[oaicite:2]{index=2}.

````python
import pandas as pd

df = pd.read_csv("dades_escola.csv")

df['Mitjana Alumne'] = df[['M01', 'M04', 'M03', 'M05']].mean(axis=1)

mitjana_tots_alumnes = df['Mitjana Alumne'].mean()

aprovats = len(df[df['Mitjana Alumne'] >= 5])
suspesos = len(df[df['Mitjana Alumne'] < 5])
percentatge_aprovats = (aprovats / len(df)) * 100
percentatge_suspesos = (suspesos / len(df)) * 100

df['Nota Mínima'] = df[['M01', 'M04', 'M03', 'M05']].min(axis=1)
df['Nota Màxima'] = df[['M01', 'M04', 'M03', 'M05']].max(axis=1)

nota_minima_total = df[['M01', 'M04', 'M03', 'M05']].min().min()
nota_maxima_total = df[['M01', 'M04', 'M03', 'M05']].max().max()

# Resultats
print("Mitjana de nota final de cada alumne:")
print(df[['Professor', 'Alumne', 'Mitjana Alumne']])

print("\nMitjana de la nota final de tots els alumnes:", mitjana_tots_alumnes)

print("\nPercentatge d'aprovats:", percentatge_aprovats, "%")
print("Percentatge de suspesos:", percentatge_suspesos, "%")

print("\nNotes més baixes i més altes de cada alumne:")
print(df[['Professor', 'Alumne', 'Nota Mínima', 'Nota Màxima']])

print("\nNota mínima global de tots els alumnes:", nota_minima_total)
print("Nota màxima global de tots els alumnes:", nota_maxima_total)
````

## Exercici 6

````python
import matplotlib.pyplot as plt
import math

# Definim una funció per calcular la mitjana de notes per alumne
def mitjana_alumne(notes):
    return sum(notes) / len(notes)

# Dades dels alumnes amb les seves notes
alumnes = {
    "Alumne 1": [7, 8, 9],
    "Alumne 2": [5, 6, 7],
    "Alumne 3": [9, 10, 8]
}

# Càlcul de la mitjana de cada alumne
mitjanes = [mitjana_alumne(notes) for notes in alumnes.values()]

# 1. Gràfic de la mitjana per alumne
plt.bar(alumnes.keys(), mitjanes)
plt.title('Mitjana per Alumne')
plt.xlabel('Alumne')
plt.ylabel('Mitjana')
plt.show()

# 2. Gràfic de la distribució de notes
totes_notes = [note for notes in alumnes.values() for note in notes]
plt.hist(totes_notes, bins=10, edgecolor='black')
plt.title('Distribució de Notes')
plt.xlabel('Nota')
plt.ylabel('Nombre de Notes')
plt.show()

# 3. Percentatge d'aprovats i suspesos
def percentatge_aprovats_suspesos(alumnes):
    aprovats = 0
    suspesos = 0
    total_alumnes = len(alumnes)

    for notes in alumnes.values():
        if mitjana_alumne(notes) >= 5:
            aprovats += 1
        else:
            suspesos += 1

    return aprovats / total_alumnes * 100, suspesos / total_alumnes * 100

aprovats, suspesos = percentatge_aprovats_suspesos(alumnes)

# 4. Gràfic de percentatges d'aprovats i suspesos
labels = ['Aprovats', 'Suspesos']
sizes = [aprovats, suspesos]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Percentatge d\'Aprovats i Suspensos')
plt.show()

# 5. Càlcul de les notes més baixes i més altes
def notes_extremes(alumnes):
    notes_totes = [note for notes in alumnes.values() for note in notes]
    return min(notes_totes), max(notes_totes)

min_note, max_note = notes_extremes(alumnes)

# 6. Gràfic de la nota més baixa i més alta
plt.plot(['Mínima', 'Màxima'], [min_note, max_note], marker='o')
plt.title('Notes Extremes')
plt.ylabel('Nota')
plt.show()

````

# Explicació del Logging

## Configuració del Logger
S'ha configurat el *logging* per escriure els missatges al fitxer `informe.log`. El format de cada missatge inclou:
- **Data i hora**: Cada missatge conté la data i hora exactes de la seva creació.
- **Nivell de severitat**: El nivell de severitat pot ser **INFO**, **WARNING**, **ERROR**, etc.
- **Missatge**: El missatge especifica el que ha passat durant l'execució del codi.

## Registre d'Informació
El registre d'informació es realitza a cada pas important del codi. Alguns exemples són:
- L'inici del procés d'anàlisi de dades.
- El càlcul de les mitjanes per alumne.
Això ajuda a fer un seguiment del progrés del programa i a identificar etapes crítiques en el codi.

## Registre d'Advertència
S'ha afegit una advertència quan no es troben alumnes en les dades. Això pot ajudar a detectar possibles problemes amb les dades d'entrada, com ara:
- Si el conjunt d'alumnes està buit o mal format.
  
## Registre d'Errors
Si hi ha un error durant l'execució del codi, com per exemple una divisió per zero, aquest es registra amb un missatge d'error específic. Alguns exemples poden ser:
- Errors en el càlcul de mitjanes si es produeix una excepció.
- Missatges detallats sobre l'error per a facilitar la depuració.
